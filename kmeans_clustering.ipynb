```python
# Import required packages
import pandas as pd
from sklearn.preprocessing import StandardScaler
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt
import seaborn as sns; sns.set()  # for plot styling
import numpy as np
import os
import glob
from kneed import DataGenerator, KneeLocator
```

```python
def kmeansclustering(df):
    
    # Create new column for standardized data 
    df['mean_distance.s'] = ''

    # Scale data within each category
    ss=StandardScaler()
    df[['mean_distance.s']]=ss.fit_transform(df[['mean_distance']])

    Sum_of_squared_distances = []
    # Make max cluster value the number of items in each category
    K = range(1,15)
    for k in K:
        km = KMeans(n_clusters=k)
        km = km.fit(df[['mean_distance.s']])
        Sum_of_squared_distances.append(km.inertia_)

    # Visualization
    plt.plot(K, Sum_of_squared_distances, 'bx-')
    plt.xlabel('k')
    plt.ylabel('Sum_of_squared_distances')
    kl = KneeLocator(range(1, 15), Sum_of_squared_distances, curve="convex", direction="decreasing")
    plt.title('The Optimal k in the ' + CatName + ' Category is ' + str(kl.elbow))
    return plt.show()
```

```python
# Read in data
path = '/Users/arianayoum/Desktop/Outside Project/Pilot_data/inputs/cleandata/*.csv'
for file in glob.glob(path):
    filename = os.path.basename(file)
    data = pd.read_csv(file)
    data.columns.values[0] = "Category"
    CatName = filename[:-4]
    data['Category'] = CatName
    kmeansclustering(data)
```
